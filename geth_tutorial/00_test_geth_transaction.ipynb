# 测试私链
env：go-ethereum / v1.9 and make geth


## 启动与初始化

```shell

# 初始化创世区块
geth --datadir "blockchain/db" init blockchain/gensis.json

# 启动私有链，并保持开启
geth --rpc --rpcport 8545 --rpccorsdomain "*" --datadir "./blockchain/db" --port 30303 --rpcapi "eth,net,web3,personal,admin,shh,txpool,debug,miner" --networkid 100000 --rpcaddr=0.0.0.0 --nodiscover --allow-insecure-unlock 2>> eth.log &

# 另一个terminal
geth attach rpc:./blockchain/db/geth.ipc

```
